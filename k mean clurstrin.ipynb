```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('income.csv')
df
```

```python
plt.scatter(df.Age, df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
```

```python
df
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]
plt.scatter(df0.Age, df0['Income($)'], color='gray')
plt.scatter(df1.Age, df1['Income($)'], color='red')
plt.scatter(df2.Age, df2['Income($)'], color='black')
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])#to make the scale on x and y axis equal(0 to 1)
df['Income($)'] = scaler.transform(df[['Income($)']])
df
scaler.fit(df[['Age']])#to make the scale on x and y axis equal(0 to 1)
df['Age'] = scaler.transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted

df
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]
plt.scatter(df0.Age, df0['Income($)'], color='gray')
plt.scatter(df1.Age, df1['Income($)'], color='red')
plt.scatter(df2.Age, df2['Income($)'], color='black')
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]
plt.scatter(df0.Age, df0['Income($)'], color='gray')
plt.scatter(df1.Age, df1['Income($)'], color='red')
plt.scatter(df2.Age, df2['Income($)'], color='black')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],color="purple", marker='*', label='centroid')
plt.legend()
```

```python
km.cluster_centers_
```

```python
#now elbow plot method
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age', 'Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of square error')
plt.plot(k_rng, sse)
```
